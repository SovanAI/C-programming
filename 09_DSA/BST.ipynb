```python
#include <stdio.h>
#include <stdlib.h>

struct Node {
    int data;
    struct Node *left;
    struct Node *right;
};

struct Node* createNode(int value) {
    struct Node *newNode = (struct Node*)malloc(sizeof(struct Node));
    newNode->data = value;
    newNode->left = NULL;
    newNode->right = NULL;
    return newNode;
}

/* Insert into BST */
struct Node* insert(struct Node *root, int value) {
    if (root == NULL)
        return createNode(value);

    if (value < root->data)
        root->left = insert(root->left, value);
    else if (value > root->data)
        root->right = insert(root->right, value);

    return root;
}

/* Search in BST */
void search(struct Node *root, int value) {
    if (root == NULL) {
        printf("Not Found?!\n");
        return;
    }
    if (root->data == value) {
        printf("Found!!\n");
        return;
    }
    if (value < root->data)
        search(root->left, value);
    else
        search(root->right, value);
}

/* Get inorder successor (minimum in right subtree) */
struct Node* get_successor(struct Node *root) {
    struct Node *curr = root->right;
    while (curr != NULL && curr->left != NULL)
        curr = curr->left;
    return curr;
}

/* Delete a node from BST */
struct Node* deleteNode(struct Node *root, int value) {
    if (root == NULL)
        return root;

    if (value < root->data) {
        root->left = deleteNode(root->left, value);
    } else if (value > root->data) {
        root->right = deleteNode(root->right, value);
    } else {
        /* Node found */
        if (root->left == NULL) {
            struct Node *temp = root->right;
            free(root);
            return temp;
        }
        if (root->right == NULL) {
            struct Node *temp = root->left;
            free(root);
            return temp;
        }

        struct Node *succ = get_successor(root);
        root->data = succ->data;
        root->right = deleteNode(root->right, succ->data);
    }
    return root;
}

/* Inorder Traversal */
void inorder_traversal(struct Node *root) {
    if (root != NULL) {
        inorder_traversal(root->left);
        printf("Inorder_traversal: %d\n", root->data);
        inorder_traversal(root->right);
    }
}

int main() {
    struct Node *root = NULL;

    root = insert(root, 10);
    root = insert(root, 5);
    root = insert(root, 15);
    root = insert(root, 3);
    root = insert(root, 60);
    root = insert(root, 49);

    printf("\n");

    search(root, 5);
    search(root, 55);

    printf("\n");
    inorder_traversal(root);

    printf("\n");
    root = deleteNode(root, 15);
    inorder_traversal(root);

    return 0;
}
```

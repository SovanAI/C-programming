```python
class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next

class SinglyLL:
    def __init__(self, head=None):
        self.head = head

    def insert_at_end(self, value):
        new_node = Node(value)

        if(self.head != None):
            temp = self.head
            while(temp.next != None):
                temp = temp.next
            temp.next = new_node
        else:
            self.head = new_node
    
    def insert_at_begining(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    def insert_at_any(self, value, find):
        new_node = Node(value)
        temp = self.head
        while temp != None:
            if (temp.data == find):
                new_node.next = temp.next
                temp.next = new_node
            temp = temp.next

    def deleteNode(self, value):
        temp = self.head
        prev = temp

        if(temp.data == value):
            self.head = temp.next

        while temp != None:
            if(temp.data == value):
                prev.next = temp.next
                break
            else:
                prev = temp
                temp = temp.next

    def PrintLL(self):
        temp = self.head
        while(temp != None):
            print(temp.data, end=" ")
            temp = temp.next


L = SinglyLL()
L.insert_at_end(10)
L.insert_at_end(20)
L.insert_at_end(30)
L.insert_at_end(40)
L.insert_at_begining(5)
L.insert_at_any(25, 20)
L.deleteNode(40)
L.deleteNode(30)
L.PrintLL()
```

```python
#include <stdio.h>
#include <stdlib.h>

struct Node {
    int data;
    struct Node *next;
};

struct SinglyLL {
    struct Node *head;
};

struct Node* createNode(int value) {
    struct Node *newNode = (struct Node*)malloc(sizeof(struct Node));
    newNode->data = value;
    newNode->next = NULL;
    return newNode;
}

void insert_at_end(struct SinglyLL *L, int value) {
    struct Node *newNode = createNode(value);

    if (L->head != NULL) {
        struct Node *temp = L->head;
        while (temp->next != NULL) {
            temp = temp->next;
        }
        temp->next = newNode;
    } else {
        L->head = newNode;
    }
}

void insert_at_beginning(struct SinglyLL *L, int value) {
    struct Node *newNode = createNode(value);
    newNode->next = L->head;
    L->head = newNode;
}

void insert_at_any(struct SinglyLL *L, int value, int find) {
    struct Node *temp = L->head;

    while (temp != NULL) {
        if (temp->data == find) {
            struct Node *newNode = createNode(value);
            newNode->next = temp->next;
            temp->next = newNode;
            return;
        }
        temp = temp->next;
    }
}

void deleteNode(struct SinglyLL *L, int value) {
    struct Node *temp = L->head;
    struct Node *prev = NULL;

    if (temp != NULL && temp->data == value) {
        L->head = temp->next;
        free(temp);
        return;
    }

    while (temp != NULL) {
        if (temp->data == value) {
            prev->next = temp->next;
            free(temp);
            return;
        }
        prev = temp;
        temp = temp->next;
    }
}

void printLL(struct SinglyLL *L) {
    struct Node *temp = L->head;
    while (temp != NULL) {
        printf("%d ", temp->data);
        temp = temp->next;
    }
    printf("\n");
}

int main() {
    struct SinglyLL L;
    L.head = NULL;

    insert_at_end(&L, 10);
    insert_at_end(&L, 20);
    insert_at_end(&L, 30);
    insert_at_end(&L, 40);
    insert_at_beginning(&L, 5);
    insert_at_any(&L, 25, 20);
    deleteNode(&L, 40);
    deleteNode(&L, 30);

    printLL(&L);

    return 0;
}
```

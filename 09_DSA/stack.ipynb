```python
class Stack:
    def __init__(self):
        self.items = []

    # Push operation
    def push(self, item):
        self.items.append(item)

    # Pop operation
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return "Stack is empty"

    # Peek operation
    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return "Stack is empty"

    # Check if stack is empty
    def is_empty(self):
        return len(self.items) == 0

    # Get size of stack
    def size(self):
        return len(self.items)

    # Display stack
    def display(self):
        return self.items


# Example usage
stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)

print("Stack:", stack.display())   # [10, 20, 30]
print("Top element:", stack.peek())  # 30
print("Popped:", stack.pop())        # 30
print("Stack after pop:", stack.display())  # [10, 20]
print("Is empty?", stack.is_empty()) # False
```

```python
#include <stdio.h>

#define MAX 100

struct Stack {
    int items[MAX];
    int top;
};

void initStack(struct Stack *s) {
    s->top = -1;
}

/* Push operation */
void push(struct Stack *s, int item) {
    if (s->top == MAX - 1) {
        printf("Stack overflow\n");
        return;
    }
    s->items[++s->top] = item;
}

/* Pop operation */
int pop(struct Stack *s) {
    if (s->top == -1) {
        printf("Stack is empty\n");
        return -1;
    }
    return s->items[s->top--];
}

/* Peek operation */
int peek(struct Stack *s) {
    if (s->top == -1) {
        printf("Stack is empty\n");
        return -1;
    }
    return s->items[s->top];
}

/* Check if stack is empty */
int is_empty(struct Stack *s) {
    return s->top == -1;
}

/* Get size of stack */
int size(struct Stack *s) {
    return s->top + 1;
}

/* Display stack */
void display(struct Stack *s) {
    if (is_empty(s)) {
        printf("Stack is empty\n");
        return;
    }

    printf("Stack: ");
    for (int i = 0; i <= s->top; i++) {
        printf("%d ", s->items[i]);
    }
    printf("\n");
}

int main() {
    struct Stack stack;
    initStack(&stack);

    push(&stack, 10);
    push(&stack, 20);
    push(&stack, 30);

    display(&stack);                     // 10 20 30
    printf("Top element: %d\n", peek(&stack));
    printf("Popped: %d\n", pop(&stack));
    display(&stack);                     // 10 20
    printf("Is empty? %s\n", is_empty(&stack) ? "True" : "False");

    return 0;
}
```

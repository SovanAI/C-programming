```python
#include <stdio.h>

#define SIZE 100

struct Queue {
    int queue[SIZE];
    int front;
    int rear;
};

void initQueue(struct Queue *q) {
    q->front = 0;
    q->rear = -1;
}

int isEmpty(struct Queue *q) {
    if (q->rear < q->front) {
        printf("Queue is Empty!\n");
        return 1;
    }
    return 0;
}

void enqueue(struct Queue *q, int value) {
    q->queue[++q->rear] = value;
}

void dequeue(struct Queue *q) {
    if (!isEmpty(q)) {
        q->front++;
    }
}

void printQueue(struct Queue *q) {
    if (isEmpty(q)) return;

    for (int i = q->front; i <= q->rear; i++) {
        printf("%d ", q->queue[i]);
    }
    printf("\n");
}

int main() {
    struct Queue Q;
    initQueue(&Q);

    /* Equivalent to Q.delete() in Python */
    dequeue(&Q);

    printQueue(&Q);

    return 0;
}
```

```python
# Double Ended Queue:
class DQueue:
    def __init__(self):
        self.queue = []

    def isEmpty(self):
        if len(self.queue) == 0:
            print("Queue is Empty!")
            return True
        return False
    
    def insert_at_end(self, value):
        self.queue.append(value)
        
    def insert_at_beg(self, value):
        self.queue.insert(0, value) 

    def delete_at_front(self):
        if not self.isEmpty():
            self.queue.pop(0)
            
    def delete_at_end(self):
        if not self.isEmpty():
            self.queue.pop()

    def print(self):
        for value in self.queue:
            print(value, end=' ')



dQ = DQueue()
dQ.insert_at_end(10)
dQ.insert_at_beg(20)
dQ.delete_at_end()
dQ.delete_at_front()
# Q.delete()
dQ.print()
```

```python
#include <stdio.h>

#define SIZE 100

struct DQueue {
    int queue[SIZE];
    int front;
    int rear;
};

void initDQueue(struct DQueue *dq) {
    dq->front = 0;
    dq->rear = -1;
}

int isEmpty(struct DQueue *dq) {
    if (dq->rear < dq->front) {
        printf("Queue is Empty!\n");
        return 1;
    }
    return 0;
}

void insert_at_end(struct DQueue *dq, int value) {
    dq->queue[++dq->rear] = value;
}

void insert_at_beg(struct DQueue *dq, int value) {
    dq->front--;
    dq->queue[dq->front] = value;
}

void delete_at_front(struct DQueue *dq) {
    if (!isEmpty(dq)) {
        dq->front++;
    }
}

void delete_at_end(struct DQueue *dq) {
    if (!isEmpty(dq)) {
        dq->rear--;
    }
}

void printDQueue(struct DQueue *dq) {
    if (isEmpty(dq)) return;

    for (int i = dq->front; i <= dq->rear; i++) {
        printf("%d ", dq->queue[i]);
    }
    printf("\n");
}

int main() {
    struct DQueue dQ;
    initDQueue(&dQ);

    insert_at_end(&dQ, 10);
    insert_at_beg(&dQ, 20);
    delete_at_end(&dQ);
    delete_at_front(&dQ);

    printDQueue(&dQ);

    return 0;
}
```

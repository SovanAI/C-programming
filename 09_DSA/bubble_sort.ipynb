```python
def bubble_sort(arr):
    # Edge case 1: Check if input is a list
    if not isinstance(arr, list):
        print("Invalid input: Expected a list")
        return

    # Edge case 2: Empty or single-element list
    if len(arr) <= 1:
        print(arr)
        return

    n = len(arr)

    for i in range(n):
        swapped = False  # Optimization for already sorted array

        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        # Edge case 3: If no swaps, array is already sorted
        if not swapped:
            break

    print(arr)
```

```python
#include <stdio.h>

/* Bubble Sort Function */
void bubbleSort(int arr[], int n) {
    // Edge Case 1: NULL array
    if (arr == NULL) {
        printf("Array is NULL\n");
        return;
    }

    // Edge Case 2: Empty or single element array
    if (n <= 1) {
        printf("Array is already sorted\n");
        return;
    }

    int i, j, temp;
    int swapped;

    for (i = 0; i < n - 1; i++) {
        swapped = 0;   // Optimization flag

        for (j = 0; j < n - 1 - i; j++) {
            if (arr[j] > arr[j + 1]) {
                // Swap
                temp = arr[j];
                arr[j] = arr[j + 1];
                arr[j + 1] = temp;
                swapped = 1;
            }
        }

        // Edge Case 3: Already sorted array
        if (swapped == 0)
            break;
    }
}

/* Print Array */
void printArray(int arr[], int n) {
    if (arr == NULL || n <= 0) {
        printf("Nothing to print\n");
        return;
    }

    for (int i = 0; i < n; i++)
        printf("%d ", arr[i]);
    printf("\n");
}

int main() {
    int arr[] = {25, 60, 5, 36, 48, 48, 5};
    int n = sizeof(arr) / sizeof(arr[0]);

    bubbleSort(arr, n);
    printArray(arr, n);

    return 0;
}
```

```python
#Adjacency List

class Graph:
    def __init__(self):
        self.adjlist = {}

    def add_vertex(self, vertex: int):
        if vertex not in self.adjlist:
            self.adjlist[vertex] = []

    def add_edge(self, src: int, dest: int):
        self.add_vertex(src)
        self.add_vertex(dest)

        self.adjlist[src].append(dest)
        self.adjlist[dest].append(src)

    def Print_Graph(self):
        for vertex in self.adjlist:
            print(vertex, " --> ", self.adjlist[vertex], end="\n")




G = Graph()

G.add_vertex(5)
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(4,3)
G.add_edge(5,2)
G.add_edge(1,5)
G.add_edge(1,4)
G.Print_Graph()

```

```python
#DFS

class Graph:
    def __init__(self, vertex):
        self.matrix = [ [0]*vertex for _ in range(vertex) ]
        self.size = vertex

    def add_edge(self, start, end):
        if (0 <= start, end <= self.size):
            self.matrix[start][end] = 1
            self.matrix[end][start] = 1
        else:
            print("Invalid Edge...😒")

    def print(self):
        for row in self.matrix:
            print(' '.join(map(str, row)))

    def dfs(self, src):
        visited = [False] * self.size
        stack = [src] #insert the first element in stack

        while(stack):
            poped = stack.pop()
            
            if(visited[poped] == False):
                print(poped, end=" --> ")
                visited[poped] = True

            #now i hvae to check that from that poped node how many nodes there is:
            for vertex in range(self.size):
                if(self.matrix[poped][vertex] == 1 and visited[vertex] == False):
                    stack.append(vertex)

G = Graph(6)

G.add_edge(0,1)
G.add_edge(0,2)
G.add_edge(2,3)
G.add_edge(2,4)
G.add_edge(3,5)
G.add_edge(4,5)

G.dfs(0)

```

```python
#include <stdio.h>

#define MAX 100

struct Graph {
    int matrix[MAX][MAX];
    int size;
};

void initGraph(struct Graph *G, int vertex) {
    G->size = vertex;

    for (int i = 0; i < vertex; i++) {
        for (int j = 0; j < vertex; j++) {
            G->matrix[i][j] = 0;
        }
    }
}

void add_edge(struct Graph *G, int start, int end) {
    if (start >= 0 && start < G->size && end >= 0 && end < G->size) {
        G->matrix[start][end] = 1;
        G->matrix[end][start] = 1;
    } else {
        printf("Invalid Edge...\n");
    }
}

void dfs(struct Graph *G, int src) {
    int visited[MAX] = {0};
    int stack[MAX];
    int top = -1;

    /* push source */
    stack[++top] = src;

    while (top != -1) {
        int popped = stack[top--];

        if (!visited[popped]) {
            printf("%d --> ", popped);
            visited[popped] = 1;
        }

        for (int vertex = G->size - 1; vertex >= 0; vertex--) {
            if (G->matrix[popped][vertex] == 1 && !visited[vertex]) {
                stack[++top] = vertex;
            }
        }
    }
}

int main() {
    struct Graph G;
    initGraph(&G, 6);

    add_edge(&G, 0, 1);
    add_edge(&G, 0, 2);
    add_edge(&G, 2, 3);
    add_edge(&G, 2, 4);
    add_edge(&G, 3, 5);
    add_edge(&G, 4, 5);

    dfs(&G, 0);

    return 0;
}
```

```python
#BFS
from collections import deque

class Graph:
    def __init__(self, vertex):
        self.matrix = [ [0]*vertex for _ in range(vertex) ]
        self.size = vertex

    def add_edge(self, start, end):
        if (0 <= start, end <= self.size):
            self.matrix[start][end] = 1
            self.matrix[end][start] = 1
        else:
            print("Invalid Edge...😒")

    def print(self):
        for row in self.matrix:
            print(' '.join(map(str, row)))

    def BFS(self, src):
        visited = [False] * self.size
        queue = deque([src]) #insert the first element in queue
        visited[src]= True
        while(queue):
            poped = queue.popleft()
            print(poped, end=" --> ")

            #now i hvae to check that from that poped node how many nodes there is:
            for vertex in range(self.size):
                if(self.matrix[src][vertex] == 1 and visited[vertex] == False):
                    queue.append(vertex)
                    visited[vertex] = True

G = Graph(6)

G.add_edge(0,1)
G.add_edge(0,2)
G.add_edge(2,3)
G.add_edge(2,4)
G.add_edge(3,5)
G.add_edge(4,5)

G.BFS(0)
```

```python
#include <stdio.h>

#define MAX 100

struct Graph {
    int matrix[MAX][MAX];
    int size;
};

void initGraph(struct Graph *G, int vertex) {
    G->size = vertex;
    for (int i = 0; i < vertex; i++) {
        for (int j = 0; j < vertex; j++) {
            G->matrix[i][j] = 0;
        }
    }
}

void add_edge(struct Graph *G, int start, int end) {
    if (start >= 0 && start < G->size && end >= 0 && end < G->size) {
        G->matrix[start][end] = 1;
        G->matrix[end][start] = 1;   // undirected graph
    } else {
        printf("Invalid Edge...\n");
    }
}

void BFS(struct Graph *G, int src) {
    int visited[MAX] = {0};
    int queue[MAX];
    int front = 0, rear = -1;

    /* enqueue source */
    queue[++rear] = src;
    visited[src] = 1;

    while (front <= rear) {
        int popped = queue[front++];
        printf("%d --> ", popped);

        for (int vertex = 0; vertex < G->size; vertex++) {
            if (G->matrix[popped][vertex] == 1 && !visited[vertex]) {
                queue[++rear] = vertex;
                visited[vertex] = 1;
            }
        }
    }
}

int main() {
    struct Graph G;
    initGraph(&G, 6);

    add_edge(&G, 0, 1);
    add_edge(&G, 0, 2);
    add_edge(&G, 2, 3);
    add_edge(&G, 2, 4);
    add_edge(&G, 3, 5);
    add_edge(&G, 4, 5);

    BFS(&G, 0);

    return 0;
}
```

```python
#include <stdio.h>
#include <stdlib.h>

#define SIZE 5

struct CircularQueue {
    int queue[SIZE];
    int front;
    int rear;
};

void initQueue(struct CircularQueue *cq) {
    cq->front = -1;
    cq->rear = -1;
}

int isFull(struct CircularQueue *cq) {
    return (cq->rear + 1) % SIZE == cq->front;
}

int isEmpty(struct CircularQueue *cq) {
    return cq->front == -1;
}

void enqueue(struct CircularQueue *cq, int value) {
    if (isFull(cq)) {
        printf("Queue is Full!!\n");
        return;
    }

    if (isEmpty(cq)) {
        cq->front = cq->rear = 0;
    } else {
        cq->rear = (cq->rear + 1) % SIZE;
    }

    cq->queue[cq->rear] = value;
}

void dequeue(struct CircularQueue *cq) {
    if (isEmpty(cq)) {
        printf("Queue is Empty!!\n");
        return;
    }

    printf("Deleted value is: %d\n", cq->queue[cq->front]);

    if (cq->front == cq->rear) {
        cq->front = cq->rear = -1;
    } else {
        cq->front = (cq->front + 1) % SIZE;
    }
}

void display(struct CircularQueue *cq) {
    if (isEmpty(cq)) {
        printf("Queue is Empty\n");
        return;
    }

    int i = cq->front;
    printf("Queue elements: ");
    while (1) {
        printf("%d ", cq->queue[i]);
        if (i == cq->rear)
            break;
        i = (i + 1) % SIZE;
    }
    printf("\n");
}

int main() {
    struct CircularQueue cq;
    initQueue(&cq);

    display(&cq);
    dequeue(&cq);
    dequeue(&cq);

    enqueue(&cq, 10);
    enqueue(&cq, 20);
    enqueue(&cq, 30);
    enqueue(&cq, 40);
    enqueue(&cq, 50);
    enqueue(&cq, 60);   // overflow case

    display(&cq);

    dequeue(&cq);
    dequeue(&cq);
    dequeue(&cq);
    dequeue(&cq);
    dequeue(&cq);
    dequeue(&cq);       // underflow case

    display(&cq);

    return 0;
}
```

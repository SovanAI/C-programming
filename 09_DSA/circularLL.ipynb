```python
class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next


class CircularSinglyLL:
    def __init__(self, head=None):
        self.head = head

    def insert_at_end(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def insert_at_begining(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            new_node.next = self.head
            temp.next = new_node
            self.head = new_node

    def insert_at_any(self, value, find):
        if self.head is None:
            return

        new_node = Node(value)
        temp = self.head

        while True:
            if temp.data == find:
                new_node.next = temp.next
                temp.next = new_node
                return
            temp = temp.next
            if temp == self.head:
                break

    def deleteNode(self, value):
        if self.head is None:
            return

        temp = self.head
        prev = None

        if self.head.data == value:
            if self.head.next == self.head:
                self.head = None
                return

            prev = self.head
            while prev.next != self.head:
                prev = prev.next
            self.head = self.head.next
            prev.next = self.head
            return

        prev = self.head
        temp = self.head.next

        while temp != self.head:
            if temp.data == value:
                prev.next = temp.next
                return
            prev = temp
            temp = temp.next

    def PrintLL(self):
        if self.head is None:
            print("List is empty")
            return

        temp = self.head
        while True:
            print(temp.data, end=" ")
            temp = temp.next
            if temp == self.head:
                break
        print()


L = CircularSinglyLL()
L.insert_at_end(10)
L.insert_at_end(20)
L.insert_at_end(30)
L.insert_at_end(40)
L.insert_at_begining(5)
L.insert_at_any(25, 20)
L.deleteNode(40)
L.deleteNode(30)
L.PrintLL()
```

```python
#include <stdio.h>
#include <stdlib.h>

struct Node {
    int data;
    struct Node *next;
};

struct CircularSLL {
    struct Node *head;
};

struct Node* createNode(int value) {
    struct Node *newNode = (struct Node*)malloc(sizeof(struct Node));
    newNode->data = value;
    newNode->next = NULL;
    return newNode;
}

void insert_at_end(struct CircularSLL *L, int value) {
    struct Node *newNode = createNode(value);

    if (L->head == NULL) {
        L->head = newNode;
        newNode->next = L->head;
        return;
    }

    struct Node *temp = L->head;
    while (temp->next != L->head) {
        temp = temp->next;
    }

    temp->next = newNode;
    newNode->next = L->head;
}

void insert_at_beginning(struct CircularSLL *L, int value) {
    struct Node *newNode = createNode(value);

    if (L->head == NULL) {
        L->head = newNode;
        newNode->next = L->head;
        return;
    }

    struct Node *temp = L->head;
    while (temp->next != L->head) {
        temp = temp->next;
    }

    newNode->next = L->head;
    temp->next = newNode;
    L->head = newNode;
}

void insert_at_any(struct CircularSLL *L, int value, int find) {
    if (L->head == NULL)
        return;

    struct Node *temp = L->head;

    do {
        if (temp->data == find) {
            struct Node *newNode = createNode(value);
            newNode->next = temp->next;
            temp->next = newNode;
            return;
        }
        temp = temp->next;
    } while (temp != L->head);
}

void deleteNode(struct CircularSLL *L, int value) {
    if (L->head == NULL)
        return;

    struct Node *temp = L->head;
    struct Node *prev = NULL;

    /* Delete head node */
    if (temp->data == value) {
        if (temp->next == temp) {
            free(temp);
            L->head = NULL;
            return;
        }

        prev = temp;
        while (prev->next != L->head) {
            prev = prev->next;
        }

        L->head = temp->next;
        prev->next = L->head;
        free(temp);
        return;
    }

    prev = L->head;
    temp = L->head->next;

    while (temp != L->head) {
        if (temp->data == value) {
            prev->next = temp->next;
            free(temp);
            return;
        }
        prev = temp;
        temp = temp->next;
    }
}

void printLL(struct CircularSLL *L) {
    if (L->head == NULL) {
        printf("List is empty\n");
        return;
    }

    struct Node *temp = L->head;
    do {
        printf("%d ", temp->data);
        temp = temp->next;
    } while (temp != L->head);

    printf("\n");
}

int main() {
    struct CircularSLL L;
    L.head = NULL;

    insert_at_end(&L, 10);
    insert_at_end(&L, 20);
    insert_at_end(&L, 30);
    insert_at_end(&L, 40);
    insert_at_beginning(&L, 5);
    insert_at_any(&L, 25, 20);
    deleteNode(&L, 40);
    deleteNode(&L, 30);

    printLL(&L);

    return 0;
}
```
